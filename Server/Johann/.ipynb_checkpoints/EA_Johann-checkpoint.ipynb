# Facturacion

```python
df_factu=pd.read_csv("/home/jovyan/Proyecto/Data/FacturacionCorregida.csv",sep=";",parse_dates=['Fecha'])
df_factu["descuento"]=(df_factu['Suma de Vlr_Bruto']-df_factu['Suma de Vlr_Subtotal'])/df_factu['Suma de Vlr_Bruto']
df_factu['Precio_final_product']=df_factu['Suma de Vlr_Subtotal']/df_factu['Cantidad inv.']
df_factu['mes']=df_factu['Fecha'].dt.month
df_factu['weekday']=df_factu['Fecha'].dt.weekday
df_factu
```

```python
df_factu.info()
```

```python
tiendas=df_factu.iloc[:,[0,1]].drop_duplicates()
tiendas
```

```python
df_factu.groupby('Desc. C.O.')['Cantidad inv.'].sum().sort_values(ascending=False).to_frame()
```

```python
df_factu.groupby('Fecha')['Cantidad inv.'].sum().plot(figsize=(20,10))
```

```python
df_factu.groupby('mes')['Cantidad inv.'].sum().plot(figsize=(10,5))
plt.title('ventas totales por mes')
```

```python
df_factu.groupby('weekday')['Cantidad inv.'].sum().plot(figsize=(10,5))
plt.title('ventas totales por semana')
```

```python
groups=['Fecha','mes','weekday']
tiendas=df_factu['Desc. C.O.'].unique()
for g in groups:
    df_aux=df_factu.groupby([g,'Desc. C.O.'])['Cantidad inv.'].sum().reset_index('Desc. C.O.')
    for t in tiendas:
        df_aux[df_aux['Desc. C.O.']==t].plot()
        plt.title('{}'.format(t))
```

<b>by store</b>

```python
df_stores_refs=df_factu.groupby(['Desc. C.O.','REF_Combinada']).aggregate(Cantidad_vendida=pd.NamedAgg(column="Cantidad inv.", aggfunc="sum"),
                                                    Subtotal=pd.NamedAgg(column="Suma de Vlr_Subtotal", aggfunc="sum"), 
                                                    Precio_promedio=pd.NamedAgg(column='Precio_final_product',aggfunc='mean'),
                                                    descuento_promedio=pd.NamedAgg(column='descuento', aggfunc= 'mean')).reset_index().merge(df_maestro, on='REF_Combinada').sort_values(by='Cantidad_vendida', ascending=False)
df_stores_refs.head(10)
```

```python
tiendas_subcategorias=pd.pivot_table(data=df_stores_refs,index='Desc. C.O.',columns='SUBCATEGORIA',values='Cantidad_vendida',aggfunc='sum')
```

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.heatmap(tiendas_subcategorias,cmap='Reds')
plt.title('número de ventas')
```

```python
tiendas_subcategorias_v=pd.pivot_table(data=df_stores_refs,index='Desc. C.O.',columns='SUBCATEGORIA',values='Subtotal',aggfunc='sum')
fig,ax=plt.subplots(figsize=(20,10))
sns.heatmap(tiendas_subcategorias_v,cmap='Reds')
plt.title('ventas en peso')
```
